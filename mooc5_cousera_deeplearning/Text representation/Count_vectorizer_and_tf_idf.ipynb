### Word Embeddings

**Định nghĩa:** Word embeddings là các vector số học biểu diễn các từ trong một không gian nhiều chiều. Chúng được huấn luyện để nắm bắt các mối quan hệ ngữ nghĩa giữa các từ.

**Đặc điểm chính:**
- **Học được ngữ nghĩa:** Word embeddings nắm bắt các ngữ nghĩa và mối quan hệ ngữ cảnh giữa các từ. Ví dụ, các từ có nghĩa tương tự sẽ có vector gần nhau trong không gian vector.
- **Sử dụng mô hình tiên tiến:** Các phương pháp phổ biến như Word2Vec, GloVe, FastText, và các mô hình dựa trên transformer như BERT.
- **Kích thước không gian vector nhỏ hơn:** Các vector thường có kích thước từ 50 đến 300 chiều, tùy thuộc vào mô hình và cài đặt.

**Ưu điểm:**
- Hiểu biết tốt về ngữ nghĩa và ngữ cảnh.
- Vector kích thước nhỏ giúp giảm chi phí tính toán.

**Nhược điểm:**
- Cần dữ liệu lớn để huấn luyện hiệu quả.
- Mất thông tin về thứ tự từ nếu chỉ xem xét các từ đơn lẻ.

### Text Vectorization

**Định nghĩa:** Text vectorization là quá trình chuyển đổi văn bản thành các vector số học thông qua các phương pháp như Bag of Words (BoW), TF-IDF, hoặc các kỹ thuật count-based khác.

**Đặc điểm chính:**
- **Dựa trên tần suất:** Các vector được tạo ra dựa trên tần suất xuất hiện của từ trong văn bản.
- **Không nắm bắt ngữ nghĩa:** Các phương pháp đơn giản như BoW không nắm bắt được ngữ nghĩa hay ngữ cảnh của từ.
- **Kích thước không gian vector lớn:** Kích thước vector phụ thuộc vào số lượng từ trong từ điển (vocabulary), có thể rất lớn đối với các tập dữ liệu lớn.

**Ưu điểm:**
- Dễ triển khai và hiểu.
- Hiệu quả với các bài toán đơn giản và dữ liệu nhỏ.

**Nhược điểm:**
- Không nắm bắt được ngữ nghĩa và ngữ cảnh.
- Kích thước vector lớn, gây khó khăn cho việc tính toán và lưu trữ.

| **Chi phí tính toán**| Thấp hơn sau khi huấn luyện             | Cao hơn do kích thước vector lớn    |

# One-Hot Encoding (OHE)
Trong kỹ thuật này, chúng tôi biểu diễn mỗi từ duy nhất trong từ vựng bằng cách đặt một mã thông báo duy nhất có giá trị 1 và phần còn lại là 0 ở các vị trí khác trong vectơ. Nói một cách đơn giản, một biểu diễn vectơ của vectơ được mã hóa one-hot biểu diễn ở dạng 1 và 0 trong đó 1 biểu diễn vị trí mà từ đó tồn tại và 0 ở mọi nơi khác.

Let’s consider the following sentence:

Sentence: I am teaching NLP in Python
A word in this sentence may be “NLP”, “Python”, “teaching”, etc.

Vì từ điển được định nghĩa là danh sách tất cả các từ duy nhất có trong câu. Vì vậy, từ điển có thể trông giống như –

Dictionary: [‘I’, ’am’, ’teaching’,’ NLP’,’ in’, ’Python’] 

Therefore, the vector representation in this format according to the above dictionary is 

Vector for NLP: [0,0,0,1,0,0] \
Vector for Python:  [0,0,0,0,0,1]

This is just a very simple method to represent a word in vector form.

## Nhược điểm của mã hóa One-hot
1. Một trong những nhược điểm của mã hóa One-hot là Kích thước của vectơ bằng với số lượng từ duy nhất trong vốn từ vựng.

2. Mã hóa one-hot không nắm bắt được mối quan hệ giữa các từ khác nhau. Do đó, nó không truyền tải thông tin về ngữ cảnh.

# Count Vectorizer
Định nghĩa: Count Vectorizer là một kỹ thuật biểu diễn văn bản bằng cách đếm số lần xuất hiện của mỗi từ trong văn bản. Mỗi từ trong từ điển sẽ có một cột trong ma trận vector, và giá trị trong ma trận là số lần từ đó xuất hiện trong tài liệu.

**Đặc điểm chính:**

- Tạo ma trận đếm từ: Biểu diễn văn bản dưới dạng ma trận mà mỗi hàng là một tài liệu và mỗi cột là một từ trong từ điển. Giá trị trong ma trận là số lần từ xuất hiện trong tài liệu.
- Không phân biệt ngữ nghĩa: Chỉ đếm số lần xuất hiện của từ, không xem xét ngữ nghĩa hay ngữ cảnh của từ.

**Ưu điểm:**
1. Đơn giản:  CountVectorizer dễ sử dụng và dễ hiểu. Nó có các tham số cụ thể và yêu cầu cấu hình tối thiểu để bắt đầu xử lý trước văn bản.
2. Tốc độ và hiệu quả:  CountVectorizer có hiệu quả tính toán và có thể xử lý các tập dữ liệu văn bản lớn với nhiều tài liệu. Nó sử dụng biểu diễn ma trận thưa thớt để tiết kiệm bộ nhớ và thời gian xử lý, đặc biệt là khi xử lý dữ liệu có nhiều chiều.
3. Tính linh hoạt:  CountVectorizer cho phép các tùy chọn mã hóa linh hoạt, bao gồm xử lý n-gram (chuỗi từ liên tiếp) và các mẫu mã thông báo tùy chỉnh. Nó cũng cung cấp cơ hội để lọc các từ dừng và kiểm soát kích thước từ vựng.
4. Kết quả có thể diễn giải:  Ma trận thuật ngữ tài liệu kết quả từ CountVectorizer cung cấp kết quả có thể diễn giải. Mỗi ô trong ma trận biểu thị số lượng hoặc tần suất của một mã thông báo trong một tài liệu cụ thể, cho phép phân tích và khám phá trực tiếp.

**Nhược điểm**:

1. Bỏ qua thông tin ngữ nghĩa : Nó coi mỗi mã thông báo là một thực thể riêng biệt và không nắm bắt được mối quan hệ ngữ nghĩa giữa các từ. Nó không xem xét ngữ cảnh hoặc ý nghĩa của từ, điều này có thể hạn chế hiệu quả của nó trong các tác vụ đòi hỏi phải hiểu ngữ nghĩa của từ.
2. Xu hướng thiên về các từ thường gặp:  Nó gán tầm quan trọng cao hơn cho các từ thường xuyên xuất hiện trong tài liệu. Do đó, các từ phổ biến như "the", "and" hoặc "is" có thể chiếm ưu thế trong không gian tính năng trong khi có khả năng bỏ qua các từ hiếm hơn nhưng có ý nghĩa hơn.
3. Thiếu chuẩn hóa:  Không xem xét độ dài tài liệu, nghĩa là tài liệu dài hơn có thể có số lượng mã thông báo cao hơn tài liệu ngắn hơn, ngay cả khi chúng thảo luận về cùng một chủ đề. Việc thiếu chuẩn hóa này có thể ảnh hưởng đến các phân tích và thuật toán cụ thể dựa trên độ dài tài liệu.
4. Thông tin hạn chế:  Chỉ nắm bắt tần suất của các mã thông báo trong tài liệu. Không xem xét thứ tự hoặc trình tự của các từ, có thể liên quan đến các tác vụ phân tích văn bản cụ thể như phân tích tình cảm hoặc mô hình hóa ngôn ngữ.



````python
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit the vectorizer to the documents and transform the documents into a document-term matrix
X = vectorizer.fit_transform(documents)

# Get the feature names (tokens)
feature_names = vectorizer.get_feature_names_out()

# Print the feature names
print(feature_names)

# Print the document-term matrix
print(X.toarray())
````

# N-grams Vectorization
**N-gram** là một kỹ thuật trong xử lý ngôn ngữ tự nhiên (NLP) để phân tách văn bản thành các chuỗi con có độ dài cố định, gọi là "gram". Độ dài của n-gram là **n**, và **n** có thể là bất kỳ số tự nhiên nào. Thông thường, n-gram được sử dụng để đại diện cho chuỗi các từ hoặc ký tự liên tiếp trong văn bản.

### Các loại N-gram:

1. **Unigram (1-gram)**: Các đơn vị là một từ duy nhất.
2. **Bigram (2-gram)**: Các đơn vị là các cặp từ liên tiếp.
3. **Trigram (3-gram)**: Các đơn vị là các nhóm ba từ liên tiếp.
4. **N-gram (n > 3)**: Các đơn vị là các nhóm n từ liên tiếp.

### Ví dụ về N-gram:

Giả sử bạn có câu văn:

```
I love machine learning.
```

- **Unigrams (1-gram)**: Các đơn vị là các từ riêng biệt.
  ```
  ['I', 'love', 'machine', 'learning']
  ```

- **Bigrams (2-gram)**: Các đơn vị là các cặp từ liên tiếp.
  ```
  ['I love', 'love machine', 'machine learning']
  ```

- **Trigrams (3-gram)**: Các đơn vị là các nhóm ba từ liên tiếp.
  ```
  ['I love machine', 'love machine learning']
  ```

- **4-grams**: Các đơn vị là các nhóm bốn từ liên tiếp.
  ```
  ['I love machine learning']
  ```

### Ứng dụng của N-gram:

1. **Mô hình dự đoán từ tiếp theo**: N-gram thường được sử dụng trong các mô hình dự đoán, như khi bạn muốn dự đoán từ tiếp theo trong một câu. Ví dụ, với bigram, nếu bạn có chuỗi "I love", mô hình có thể dự đoán từ tiếp theo là "machine".
  
2. **Tăng cường khả năng hiểu ngữ cảnh**: Sử dụng n-gram giúp mô hình hiểu được ngữ cảnh xung quanh các từ, thay vì chỉ sử dụng các từ riêng lẻ. Điều này giúp cải thiện độ chính xác trong việc xử lý ngữ nghĩa.

3. **Phát hiện các mẫu trong văn bản**: N-gram có thể được sử dụng để nhận diện các cụm từ, mẫu câu, hoặc cấu trúc ngữ pháp thường xuyên trong văn bản.

### Ví dụ Python với N-gram:

Dưới đây là ví dụ sử dụng **nltk** (Natural Language Toolkit) trong Python để tạo n-gram:

```python
import nltk
from nltk.util import ngrams

# Câu văn
text = "I love machine learning."

# Tách từ (tokenization)
words = nltk.word_tokenize(text)

# Tạo bigram
bigrams = list(ngrams(words, 2))
print("Bigrams:", bigrams)

# Tạo trigram
trigrams = list(ngrams(words, 3))
print("Trigrams:", trigrams)
```

**Output:**
```
Bigrams: [('I', 'love'), ('love', 'machine'), ('machine', 'learning')]
Trigrams: [('I', 'love', 'machine'), ('love', 'machine', 'learning')]
```

Ở ví dụ này, `ngrams(words, 2)` tạo ra các bigram và `ngrams(words, 3)` tạo ra các trigram từ câu "I love machine learning."

### Tóm tắt:
- **N-gram** là một phương pháp chia văn bản thành các nhóm từ liên tiếp, giúp mô hình hiểu rõ hơn về ngữ cảnh và cấu trúc của văn bản.
- **Unigram, bigram, trigram** là các ví dụ phổ biến của n-gram với n = 1, 2, 3, nhưng bạn có thể sử dụng bất kỳ giá trị n nào tùy thuộc vào nhu cầu.

````python
import nltk
from nltk.util import ngrams

# Câu văn
text = "I love machine learning."

# Tách từ (tokenization)
words = nltk.word_tokenize(text)

# Tạo bigram
bigrams = list(ngrams(words, 2))
print("Bigrams:", bigrams)

# Tạo trigram
trigrams = list(ngrams(words, 3))
print("Trigrams:", trigrams)
````

## Giá trị N trong phương pháp N-gram tạo ra sự đánh đổi như thế nào?
Sự đánh đổi được tạo ra khi chọn các giá trị N. Nếu chúng ta chọn N là một số nhỏ hơn, thì nó có thể không đủ để cung cấp thông tin hữu ích nhất. Nhưng ngược lại, nếu chúng ta chọn N là một giá trị cao, thì nó sẽ tạo ra một ma trận khổng lồ với nhiều tính năng. Do đó, N-gram có thể là một kỹ thuật mạnh mẽ, nhưng nó cần được chăm sóc nhiều hơn một chút.

## Nhược điểm của N-Gram
1. Nó có quá nhiều tính năng.

2. Do có quá nhiều tính năng, bộ tính năng trở nên quá thưa thớt và tốn kém về mặt tính toán.

3. Việc chọn giá trị tối ưu của N không phải là nhiệm vụ dễ dàng.

# TfidfVectorizer
### TfidfVectorizer là gì?
TfidfVectorizer là viết tắt của “Term Frequency-Inverse Document Frequency Vectorizer.” Nó xây dựng dựa trên khái niệm của CountVectorizer nhưng kết hợp với lược đồ trọng số TF-IDF(là cải tiến của count vectorize). TF-IDF là một số liệu thống kê phản ánh tầm quan trọng của một thuật ngữ (token) trong một tài liệu trong một ngữ liệu lớn hơn.

Giá trị TF-IDF cho một thuật ngữ trong tài liệu được tính bằng cách nhân các thành phần tần suất thuật ngữ (TF) và tần suất tài liệu nghịch đảo (IDF):

- Tần suất thuật ngữ (TF) biểu thị tần suất của một thuật ngữ trong một tài liệu. Nó thường được tính bằng số lượng thuật ngữ trong tài liệu chia cho tổng số thuật ngữ trong tài liệu.
- Tần suất tài liệu nghịch đảo (IDF) đo lường độ hiếm của một thuật ngữ trong ngữ liệu. Nó được tính bằng logarit của tổng số tài liệu chia cho số tài liệu có chứa thuật ngữ đó.



TfidfVectorizer mã hóa văn bản, đếm tần suất thuật ngữ và áp dụng phép biến đổi IDF để có được biểu diễn TF-IDF. Nó tạo ra một ma trận trong đó các hàng biểu diễn các tài liệu và các cột biểu diễn các mã thông báo. Các giá trị ô biểu thị trọng số TF-IDF của mỗi mã thông báo trong mỗi tài liệu.

````python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Sample documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# CountVectorizer
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(documents)

# TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(documents)

# Get the feature names (tokens)
feature_names_count = count_vectorizer.get_feature_names_out()
feature_names_tfidf = tfidf_vectorizer.get_feature_names_out()

# Print the feature names
print("CountVectorizer feature names:", feature_names_count)
print("TfidfVectorizer feature names:", feature_names_tfidf)

# Print the document-term matrices
print("CountVectorizer document-term matrix:")
print(X_count.toarray())

print("TfidfVectorizer document-term matrix:")
print(X_tfidf.toarray())
````

Trong ma trận thuật ngữ tài liệu CountVectorizer, các giá trị biểu thị tần suất của từng mã thông báo trong mỗi tài liệu. Ví dụ, giá trị 2 ở hàng thứ hai và cột thứ hai cho biết mã thông báo “tài liệu” xuất hiện hai lần trong tài liệu thứ hai.

Trong ma trận thuật ngữ tài liệu TfidfVectorizer, các giá trị biểu thị trọng số TF-IDF của mỗi mã thông báo trong mỗi tài liệu. Trọng số TF-IDF kết hợp tần suất thuật ngữ (TF) và tần suất tài liệu nghịch đảo (IDF). Các mã thông báo thường xuyên hơn trong một tài liệu cụ thể và hiếm trong toàn bộ ngữ liệu có xu hướng có trọng số cao hơn. Ví dụ: giá trị 0,8198743 ở hàng thứ hai và cột thứ hai biểu thị trọng số cao hơn cho mã thông báo "tài liệu" trong tài liệu thứ hai so với các mã thông báo khác.

Trong khi CountVectorizer chỉ tập trung vào tần suất mã thông báo, TfidfVectorizer xem xét tần suất và độ hiếm của mã thông báo bằng cách sử dụng lược đồ trọng số TF-IDF. TfidfVectorizer thường được sử dụng để nhấn mạnh tầm quan trọng của các từ hiếm và giảm thiểu ảnh hưởng của các từ phổ biến trong bộ sưu tập tài liệu.
